Question-1

You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.

Input:
    10

   /   \

 2      7
/ \
8 4
Output:
    8

  /   \

4     10
/ \
2 7



```python
class TreeNode:

    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def convert_to_binary_search_tree(root):

    if root is None:
        return None

    def inorder_traversal(node):
        if node is not None:
            inorder_traversal(node.left)
            # Print the node's value
            print(node.val)
            inorder_traversal(node.right)

    inorder_traversal(root)

    return root

# time complexity: O(n)
# space complexity: O(n)
```

Question-2:

Given a Binary Search Tree with all unique values and two keys. Find the distance between two nodes in BST. The given keys always exist in BST.

Example:

Input-1:
n = 9
values = [8, 3, 1, 6, 4, 7, 10, 14,13]
node-1 = 6
node-2 = 14
Output-1:
The distance between the two keys = 4

Input-2:
n = 9
values = [8, 3, 1, 6, 4, 7, 10, 14,13]
node-1 = 3
node-2 = 4
Output-2:
The distance between the two keys = 2

```python
class TreeNode:

    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def find_distance_between_nodes(root, node1, node2):
    """Find the distance between two nodes in a binary search tree."""

    if root is None:
        return -1

    if root.val == node1.val or root.val == node2.val:
        return get_distance_from_root(root, node1) + get_distance_from_root(root, node2)

    left_distance = find_distance_between_nodes(root.left, node1, node2)
    right_distance = find_distance_between_nodes(root.right, node1, node2)

    if left_distance == -1 and right_distance == -1:
        return -1
    elif left_distance != -1 and right_distance != -1:
        return min(left_distance, right_distance) + 2
    else:
        return left_distance if left_distance != -1 else right_distance

def get_distance_from_root(root, node):
    """Get the distance of a node from the root."""

    if root is None:
        return -1

    if root.val == node.val:
        return 0
    elif root.val < node.val:
        return get_distance_from_root(root.right, node) + 1
    else:
        return get_distance_from_root(root.left, node) + 1

# time complexity: O(n)
# space complexity: O(h), h = height of the binary tree
```

Question-3:

Write a program to convert a binary tree to a doubly linked list.

Input:

  10

 /   \

5     20

   /   \

   30     35

Output:
5 10 30 20 35

```python
class TreeNode:

    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def convert_binary_tree_to_doubly_linked_list(root):

    if root is None:
        return None

    current = root
    prev = None

    while current is not None:
        next = current.right

        # Set the previous pointer of the current node.
        current.right = prev

        # if the previous node is not None, set its left pointer to the current node.
        if prev is not None:
            prev.left = current

        prev = current
        current = next

    # Set the left pointer of the last node to None.
    if prev is not None:
        prev.left = None

    return prev

# time complexity: O(n)
# space complexity: O(1)
```

Question-4:

Write a program to connect nodes at the same level.

Input:

   1

  /   \

 2      3

/ \    / \

4 5    6 7

Output:
1 → -1
2 → 3
3 → -1
4 → 5
5 → 6
6 → 7
7 → -1

```python
class Node:

    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        self.next = None

def connect_nodes_at_same_level(root: Node) -> Node:

    if root is None:
        return None

    # Initialize a queue to store the nodes of the binary tree.
    queue = [root]

    # Initialize a pointer to store the previous node.
    prev = None

    while queue:
        current = queue.pop(0)

        # Set the next pointer of the current node to the previous node.
        current.next = prev

        # Set the previous node to the current node.
        prev = current

        # Add the left and right children of the current node to the queue.
        if current.left is not None:
            queue.append(current.left)
        if current.right is not None:
            queue.append(current.right)

        # Set the next pointer of the leftmost node of the next level to None.
        if queue and not queue[0].left:
            queue[0].next = None

    return root

# time complexity: O(n)
# space complexity: O(n)
```
