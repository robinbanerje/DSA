Question-1:

Given preorder of a binary tree, calculate its depth (or height) [starting from depth 0]. The preorder is given as a string with two possible characters.
1.	‘l’ denotes the leaf
2.	‘n’ denotes internal node

The given tree can be seen as a full binary tree where every node has 0 or two children. The two children of a node can ‘n’ or ‘l’ or mix of both.

Examples :

Input : nlnll
Output : 2
Explanation :

image.png

Input : nlnnlll
Output : 3
Explaination:

image.png

```python
def calculate_depth_of_binary_tree(preorder):

  depth = 0
  for i in range(len(preorder)):
    if preorder[i] == 'n':
      depth += 1
    else:
      # A leaf node does not contribute to the depth of the tree.
      continue
  return depth

# time complexity: O(N), N = length of the preorder traversal
# space complexity: O(1)
```

Question-2:

Given a Binary tree, the task is to print the left view of the Binary Tree. The left view of a Binary Tree is a set of leftmost nodes for every level.

Examples:

Input:

        4

      /   \

    5     2

         /   \

        3     1

       /  \

      6    7

Output: 4 5 3 6

Explanation:

                1

            2       3

             \

               4

                 \

                    5

                       \

                         6

Output: 1 2 4 5 6

```python
def left_view(root):

  if not root:
    return

  queue = [root]
  while queue:
    node = queue.pop(0)
    print(node.data)
    if node.left:
      queue.append(node.left)
    else:
      queue.append(None)

# time complexity: O(N), N = number of nodes in the binary tree
# space complexity: O(H), H = height of the binary tree
```

Question-3:

Given a Binary Tree, print the Right view of it.
The right view of a Binary Tree is a set of nodes visible when the tree is visited from the Right side.

Examples:

Input:

             1
           /   \
          2     3
        /  \   /  \
       4   5  6    7
              \
               8

Output: Right view of the tree is 1 3 7 8

Input:

          1
         /
       8
      /
     7

Output: Right view of the tree is 1 8 7

```python
def rightView(root):

    if root is None:
        return

    queue = []  # Create a queue for level order traversal
    queue.append(root)  # Enqueue the root node

    while queue:
        size = len(queue)

        # Traverse all nodes at the current level
        for i in range(size):
            node = queue.pop(0)

            # Print the last node at each level
            if i == size - 1:
                print(node.data, end=' ')

            # Enqueue the left and right child nodes
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

# time complexity: O(n)
# space complexity is O(m), where m is the maximum number of nodes at any level in the binary tree
# maximum number of nodes at any level is approximately n/2, so space complexity = O(n)
```

Question-4:

Given a Binary Tree, The task is to print the bottom view from left to right. A node x is there in output if x is the bottommost node at its horizontal distance. The horizontal distance of the left child of a node x is equal to a horizontal distance of x minus 1, and that of a right child is the horizontal distance of x plus 1.

Examples:

Input:

                 20
               /    \
             8       22
           /   \      \
          5     3      25
               /  \
             10    14

Output: 5, 10, 3, 14, 25.

Input:

                 20
               /    \
              8      22
            /   \   /   \
           5    3  4     25
               / \
             10   14

Output: 5 10 4 14 25.

Explanation:
If there are multiple bottom-most nodes for a horizontal distance from the root, then print the later one in the level traversal.
3 and 4 are both the bottom-most nodes at a horizontal distance of 0, we need to print 4.

```python
class Node:
    def __init__(self, value):
        self.data = value
        self.left = None
        self.right = None
        self.hd = 0

def bottomView(root):
    if not root:
        return

    # Dictionary to store the bottom view nodes at each horizontal distance
    bottom_view = {}

    # Create a queue for level order traversal
    queue = deque()
    queue.append((root, 0))

    while queue:
        node, hd = queue.popleft()

        # Update the bottom view dictionary with the current node value
        bottom_view[hd] = node.data

        # Enqueue the left and right child nodes with updated horizontal distances
        if node.left:
            queue.append((node.left, hd - 1))

        if node.right:
            queue.append((node.right, hd + 1))

    # Print the bottom view nodes in the order of their horizontal distances
    for hd in sorted(bottom_view):
        print(bottom_view[hd], end=' ')

# time complexity: O(n)
# space complexity: O(n)
```
